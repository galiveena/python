```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,7))
plt.plot(df['sales'],marker='o',color='red',markerfacecolor='b',markeredgecolor='yellow',markeredgewidth=2,linewidth=15,alpha=0.9)
plt.xlabel("index")
plt.ylabel("sales")
plt.title("sales vs index")

```
